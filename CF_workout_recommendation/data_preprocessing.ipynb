```python
import re

# Compact keyword -> BodyPart mapping
BODY_PART_KEYWORDS = {
    # Core
    **dict.fromkeys(['abs', 'core', 'abdominal', 'six pack', 'oblique'], 'Abdominals'),
    # Legs
    **dict.fromkeys(['quad', 'quads', 'leg', 'legs', 'thigh', 'squat', 'lunge', 'lower'], 'Quadriceps'),
    **dict.fromkeys(['hamstring', 'hamstrings', 'deadlift', 'posterior'], 'Hamstrings'),
    **dict.fromkeys(['glute', 'glutes', 'butt'], 'Glutes'),
    **dict.fromkeys(['calf', 'calves'], 'Calves'),
    # Upper body
    **dict.fromkeys(['chest', 'pec', 'pecs', 'bench', 'push up', 'push-up', 'push'], 'Chest'),
    **dict.fromkeys(['lats', 'lat', 'pull up', 'pull-up', 'pull'], 'Lats'),
    **dict.fromkeys(['upper back', 'middle back', 'mid back', 'back', 'row'], 'Middle Back'),
    **dict.fromkeys(['lower back'], 'Lower Back'),
    **dict.fromkeys(['shoulder', 'shoulders', 'delts', 'deltoid', 'press', 'overhead', 'upper'], 'Shoulders'),
    # Arms
    **dict.fromkeys(['bicep', 'biceps', 'curl'], 'Biceps'),
    **dict.fromkeys(['tricep', 'triceps', 'dip', 'skullcrusher'], 'Triceps'),
    **dict.fromkeys(['forearm', 'forearms'], 'Forearms'),
    # Other
    **dict.fromkeys(['trap', 'traps', 'shrug'], 'Traps'),
    **dict.fromkeys(['neck'], 'Neck'),
    **dict.fromkeys(['adductor', 'adductors'], 'Adductors'),
    **dict.fromkeys(['abductor', 'abductors'], 'Abductors'),
}

def infer_bodyparts_from_text(text: str):
    if pd.isna(text):
        return set()
    t = text.lower()
    found = set()
    for kw, bp in BODY_PART_KEYWORDS.items():
        # Use word boundaries to avoid partial matches
        pattern = r'\b' + re.escape(kw) + r'\b'
        if re.search(pattern, t):
            found.add(bp)
    return found

def infer_program_bodyparts(row):
    title = row.get('title', '')
    desc = row.get('description', '')
    parts = infer_bodyparts_from_text(str(title)) | infer_bodyparts_from_text(str(desc))
    return sorted(parts)

program_df['target_bodyparts'] = program_df.apply(infer_program_bodyparts, axis=1)
program_df[['title', 'description','target_bodyparts']].sample(50)
```

### Phase 2 mapping - based on level, equipment, and bodyparts, to map exercises to workout program

```python
import ast

def parse_list_column(series):
    parsed = []
    for val in series.fillna('[]'):
        try:
            if isinstance(val, list):
                parsed.append(val)
            else:
                parsed.append(ast.literal_eval(val))
        except (ValueError, SyntaxError):
            parsed.append([val] if val not in [None, 'nan', ''] else [])
    return parsed

# Parse level column
if 'level_list' not in program_df_1.columns:
    program_df_1['level_list'] = parse_list_column(program_df_1['level'])

program_df_1[['title', 'level_list', 'equipment', 'target_bodyparts']].head()
```

```python
# Level compatibility: map exercise Level to program level_list
def exercise_level_to_prog_labels(ex_level: str):
    """Map exercise Level (Beginner/Intermediate/Expert) to program level labels."""
    if pd.isna(ex_level):
        return []
    ex_level = str(ex_level).strip()
    if ex_level == 'Beginner':
        return ['Beginner', 'Novice']
    elif ex_level == 'Intermediate':
        return ['Intermediate']
    elif ex_level == 'Expert':
        return ['Advanced']
    return []

def level_compatible(program_levels, exercise_level):
    """Check if exercise level is compatible with program levels."""
    pl = set(program_levels or [])
    if not pl:  # if no program levels specified, accept all
        return True
    needed = set(exercise_level_to_prog_labels(exercise_level))
    if not needed:  # if exercise level unknown, accept
        return True
    return len(pl & needed) > 0

# Equipment compatibility
def equipment_compatible(program_equipment, exercise_equipment):
    """Check if exercise equipment is compatible with program equipment."""
    if pd.isna(program_equipment):
        return True
    
    pe = str(program_equipment).strip()
    ee = str(exercise_equipment).strip()
    
    # Full Gym and Garage Gym accept all equipment
    if pe in ['Full Gym', 'Garage Gym']:
        return True
    
    # Dumbbell Only accepts: Dumbbell, Kettlebells, Body Only
    if pe == 'Dumbbell Only':
        return ee in ['Dumbbell', 'Kettlebells', 'Body Only', 'None']
    
    # At Home accepts: Body Only, Bands, Exercise Ball, Foam Roll, Dumbbell
    if pe == 'At Home':
        return ee in ['Body Only', 'Bands', 'None', 'Exercise Ball', 'Foam Roll', 'Dumbbell']
    
    return True
```

```python
# Build program -> exercises mapping
program_exercise_mapping = []

for idx, prow in program_df_1.iterrows():
    p_id = idx
    p_title = prow['title']
    p_levels = prow.get('level_list', []) or []
    p_equip = prow['equipment']
    p_bodyparts = set(prow.get('target_bodyparts', []))
    
    # Filter exercises by body parts
    candidates = exercise_df[exercise_df['BodyPart'].isin(p_bodyparts)].copy()
    
    # Filter by level compatibility
    candidates = candidates[
        candidates['Level'].apply(lambda lv: level_compatible(p_levels, lv))
    ]
    
    # Filter by equipment compatibility
    candidates = candidates[
        candidates['Equipment'].apply(lambda eq: equipment_compatible(p_equip, eq))
    ]
    
    # Collect exercise info
    for _, ex_row in candidates.iterrows():
        program_exercise_mapping.append({
            'program_id': p_id,
            'program_title': p_title,
            'program_description': prow['description'],
            'program_goal': prow.get('goal', ''),
            'program_level': ', '.join(p_levels),
            'program_equipment': p_equip,
            'program_bodyparts': ', '.join(sorted(p_bodyparts)),
            'exercise_id': int(ex_row['Unnamed: 0']),
            'exercise_title': ex_row['Title'],
            'exercise_bodypart': ex_row['BodyPart'],
            'exercise_equipment': ex_row['Equipment'],
            'exercise_level': ex_row['Level'],
            'exercise_rating': ex_row['Rating'] if pd.notna(ex_row['Rating']) else 0.0
        })
    
    if idx % 100 == 0:
        print(f"Processed {idx} programs...")

program_exercise_df = pd.DataFrame(program_exercise_mapping)
print(f"\nCreated {len(program_exercise_df):,} program-exercise mappings")
program_exercise_df.head(5)
```
